```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

def scrape_amazon_mobiles(base_url, pages=10):
    all_data = []
    for page in range(1, pages + 1):
        url = f"{base_url}&page={page}"
        print(f"Scraping page: {page}")

        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.82 Safari/537.36'
        }
        
        try:
            response = requests.get(url, headers=headers)
            response.raise_for_status()  # Check for request errors

            soup = BeautifulSoup(response.content, 'html.parser')

            products = soup.find_all('div', class_='s-main-slot s-result-list s-search-results sg-row')
            for product in products:
                items = product.find_all('div', class_='s-result-item')
                
                for item in items:
                    name_tag = item.find('span', class_='a-text-normal')
                    price_tag = item.find('span', class_='a-price-whole')
                    availability_tag = item.find('span', class_='a-declarative')

                    name = name_tag.text if name_tag else 'N/A'
                    price = price_tag.text if price_tag else 'N/A'
                    availability = 'Available in Amazon' if availability_tag else 'Not Available in Amazon'

                    all_data.append({
                        'Name': name,
                        'Price': price,
                        'Availability': availability
                    })

            time.sleep(2)  # Sleep to avoid hitting the server too hard

        except Exception as e:
            print(f"Error on page {page}: {e}")
            continue

    return pd.DataFrame(all_data)

# Base URL for Amazon mobile search
base_url = 'https://www.amazon.in/s?k=iphone&i=electronics&rh=n%3A976419031%2Cp_123%3A110955&dc&page=2&crid=YIRODOCBTKYJ&qid=1724491581&rnid=91049095031&sprefix=iph%2Celectronics%2C194&ref=sr_pg_1'

# Scrape data from 10 pages
df_mobiles = scrape_amazon_mobiles(base_url, pages=10)

# Save data to CSV
df_mobiles.to_csv('amazon_mobiles.csv', index=False)

print("Scraping completed and data saved to amazon_mobiles.csv")

```
